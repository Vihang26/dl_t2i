```python
import torch
from diffusers import StableDiffusionPipeline, DPMSolverMultistepScheduler
```

```python
import os
folder_path = '/content/gdrive/MyDrive/NLP/T2I/level_3_images_sd2_1'
os.makedirs(folder_path, exist_ok=True)
```

```python
model_id = "stabilityai/stable-diffusion-2-1"

# Use the DPMSolverMultistepScheduler (DPM-Solver++) scheduler here instead
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")
pipe.enable_attention_slicing()
```

```python
# Open the file in read mode
level1_gen_file = '/content/gdrive/MyDrive/NLP/T2I/level_3_gen.txt'
count = -1
with open(level1_gen_file, 'r') as file:
    for line in file:
        prompt = line.strip()
        count += 1
        image = pipe(prompt).images[0]
        image.save(f"/content/gdrive/MyDrive/NLP/T2I/level_3_images_sd2_1/{prompt}_{str(count).zfill(6)}.png")
```
